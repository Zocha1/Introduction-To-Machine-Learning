## 2. Sieci neuronowe

###**Definicja i historia sieci neuronowych**

**Sieci neuronowe** to modele obliczeniowe inspirowane biologicznymi sieciami neuronowymi, które występują w mózgach. Zostały one zaprojektowane w celu naśladowania sposobu, w jaki ludzki mózg przetwarza informacje. Idea sieci neuronowych została przedstawiona w **1943** roku przez neurofizjologa **Warrena McCullocha** i matematyka **Waltera Pittsa**., jednak prawdziwy rozwój technologii nastąpił w ostatnich dekadach dzięki postępom w dziedzinie mocy obliczeniowej i dostępności dużych zbiorów danych.

###**Architektura sieci neuronowych**
Kluczowe dla funkcjonowania sieci neuronowych są ich hiperparametry. Komponety takie jak liczba neuronów, warstw, wagi oraz bias tworzą sieci i umożliwiają sieci efektywne uczenie się i adaptację do nowych zadań.

####**1. Neurony**
**Neurony** są podstawowymi jednostkami przetwarzającymi w sieciach neuronowych. Każdy neuron odbiera dane wejściowe, przetwarza je przez funkcję aktywacji i generuje dane wyjściowe, które mogą być przekazywane do innych neuronów. Funkcje aktywacji, takie jak sigmoidalna, ReLU czy tangens hiperboliczny, decydują, czy i jak silny sygnał zostanie przekazany dalej.

####**2. Warstwy**
Sieci neuronowe składają się z różnych warstw neuronów, każda z nich pełni określoną funkcję w przetwarzaniu danych. Wasrtwy dzielimy na 3 kategorie, a każda z nich odpowiada za coś innego.


image.png

[Źródło](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcR1j7ff9GV38J_vaGxJtGU0OQ7AtkFQYsVmmDV0uGv3Zg&s)

**Warstwa wejściowa** jest pierwszą warstwą, z którą mają kontakt surowe dane z zewnętrznych źródeł. Neurony tej warstwy nie wykonują żadnych obliczeń, lecz służą głównie jako punkt przekazywania danych do głębszych warstw sieci, które znajdują się dalej w strukturze modelu.

**Warstwy ukryte** to serce sieci neuronowej, gdzie odbywa się większość przetwarzania danych. Każda warstwa ukryta składa się z neuronów, które otrzymują dane z poprzedniej warstwy, przetwarzają je przez zastosowanie funkcji aktywacyjnych, a następnie przekazują wyniki do kolejnej warstwy. Liczba warstw ukrytych i liczba neuronów w każdej z nich może się różnić, w zależności od złożoności zadania i głębokości modelu. Te warstwy umożliwiają sieci neuronowej naukę reprezentacji danych na różnych poziomach abstrakcji.

**Warstwa wyjściowa** stanowi finalny etap przetwarzania informacji, gdzie wyniki pracy poprzednich warstw są kompilowane w końcowy wynik. W zależności od typu zadania, które sieć ma realizować (np. klasyfikacja, regresja), warstwa wyjściowa może mieć różną liczbę neuronów oraz stosować odpowiednie funkcje aktywacyjne. Na przykład, w zadaniach klasyfikacyjnych często stosuje się funkcję aktywacyjną softmax, która przypisuje prawdopodobieństwo do każdej z możliwych klas.

####**3. Wagi**
Wagi są kluczowymi elementami w sieciach neuronowych, odgrywając centralną rolę w sposobie, w jaki sieć przetwarza informacje i uczy się na podstawie danych. Każde połączenie między neuronami w sieci ma przypisaną wagę, która moduluje siłę sygnału przekazywanego między neuronami.



image.png

[Źródło](https://nauka.metodolog.pl/sieci-neuronowe-zaawansowana-analiza-statystyczna/budowa-pojedynczego-nauronu_siec-neuronowa_rys-1/)

Wagi zazwyczaj inicjuje się losowo na początku procesu uczenia, aby umożliwić sieci start z różnorodnych punktów w przestrzeni rozwiązań. Dobór początkowych wartości wag może znacząco wpływać na skuteczność i szybkość uczenia sieci.
 Podczas treningu, wagi są dostosowywane za pomocą procesu zwanego propagacją wsteczną. Celem tego procesu jest minimalizacja błędu między przewidywaniami sieci a rzeczywistymi wartościami, co jest realizowane przez modyfikację wag w kierunku, który redukuje ten błąd.

####**4. Bias**
Bias to dodatkowy parametr w sieci neuronowej, używany w każdym neuronie oprócz wag. Dodaje on stałą wartość do sumy ważonej wejść, co pozwala przesunąć funkcję aktywacji w lewo lub w prawo, co może być niezbędne do lepszego dopasowania modelu do danych. Obecność biasu umożliwia neuronowi dostarczanie wyjścia różnego od zera nawet wtedy, gdy wszystkie wejścia są zerami, co zwiększa elastyczność i zdolność modelu do nauki.

####**5. Funkcja aktywacyjna**

Funkcje aktywacyjne decydują, czy neuron powinien być aktywowany, czyli przekazywać informacje dalej. Są one niezbędne do wprowadzenia nieliniowości do modelu, co pozwala sieciom neuronowym na uczenie się i modelowanie bardziej skomplikowanych wzorców w danych.

Popularne funkcje aktywacyjne:
- Sigmoid: Przekształca wartości do zakresu od 0 do 1, jest przydatna w modelach, gdzie potrzebujemy prawdopodobieństwa lub binarnej klasyfikacji.
- ReLU (Rectified Linear Unit): Zwraca 0 dla wszystkich negatywnych wartości, a dla pozytywnych zwraca wartość bez zmian. Jest to obecnie najczęściej używana funkcja ze względu na prostotę obliczeniową i dobre wyniki w praktyce.
- Tangens hiperboliczny (tanh): Przekształca wartości do zakresu od -1 do 1, co czyni go bardziej elastycznym niż sigmoid.
- Softmax: stosowana głównie w ostatniej warstwie sieci neuronowych dla zadań klasyfikacji wieloklasowej. Przekształca ona wektor liczb rzeczywistych na wektor prawdopodobieństw, które reprezentują przynależność do danej klasy.

### **Proces działania sieci neuronowej**

####**1. Propagacja wprzód**
W tej fazie, dane wejściowe są przekazywane od warstwy wejściowej przez ukryte warstwy do warstwy wyjściowej. W każdej warstwie, dane są modyfikowane przez wagi i funkcje aktywacji, co prowadzi do generacji wyniku końcowego.

####**2. Funkcja straty**
Funkcja straty mierzy, jak daleko wynik sieci neuronowej jest od oczekiwanego wyniku. Jest to kluczowy element, który informuje sieć o błędach w jej działaniu.

####**3. Propagacja wsteczna**
To mechanizm uczenia się sieci, polegający na propagacji błędu z powrotem przez sieć, od warstwy wyjściowej do warstwy wejściowej. Umożliwia to dostosowywanie wag w taki sposób, aby z czasem błąd był coraz mniejszy.

####**4. Optymalizacja**
Proces optymalizacji polega na wyborze i stosowaniu algorytmu, który skutecznie zmienia wagi w sieci. Popularne algorytmy to metody gradientu prostego, takie jak SGD, Adam czy RMSprop, które różnią się sposobami dostosowania wag na podstawie obliczonego gradientu.

####**5. Znaczenie funkcji aktywacyjnej**
Funkcje aktywacyjne nie tylko wprowadzają nieliniowość, ale również pomagają kontrolować przepływ danych w sieci podczas fazy propagacji wprzód. Bez nich, sieć neuronowa byłaby tylko prostym liniowym regresorem, niezdolnym do rozwiązania złożonych problemów klasyfikacyjnych czy regresyjnych. Funkcje te mają bezpośredni wpływ na to, jak sieć uczy się i jakie wzorce może wychwycić z danych.

## 3. [Tensorflow](https://www.tensorflow.org/), [Keras](https://keras.io/getting_started/)

**Keras** jest biblioteką stworzoną w 2015 roku. Jest to biblioteka sieci neuronowych, która została stworzona w celu szybkiego eksperymentowania z głębokimi sieciami neuronowymi. Oferuje proste i spójne interfejsy API, upraszcza wiele kroków w tworzeniu struktrury uczenia głębokiego oraz zapewnia jasne i proste komunikaty o błędach. Nie obsługuje ona natomiast obliczeń niskiego poziomu. Działa ona jako interfejs dla innych bibliotek, które stanowią swego rodzaju backend.

W 2017 Keras został zintegrowany z **TensorFlow**. TensorFlow to symboliczna biblioteka matematyczna opracowana przez Google używana w sieciach neuronowych, która najlepiej nadaje się do programowania przepływu danych w szeregu zadań. Oferuje wiele poziomów abstrakcji do budowania i uczenia modeli. Jest to bardzo rozbudowana biblioteka, która cechuje się wysoką wydajnością i skalowalnością.
