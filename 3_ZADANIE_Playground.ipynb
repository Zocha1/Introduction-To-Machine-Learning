##Ustawienia hiperparametrów

**Learing rate** jest jednym z najważniejszych hiperparametrów. Określa, jak dużą zmianę dokonujemy na wagach sieci przy każdej aktualizacji podczas procesu uczenia. Jeśli jest za mała, uczenie może trwać bardzo długo. Jeśli jest za duża, możemy przeskoczyć minimum funkcji kosztu, co prowadzi do niestabilnego uczenia lub nawet dywergencji.

**Activation** ustawia funkcję aktywacji, które pomagają decydować, czy neuron powinien być aktywowany czy nie. Dostępne opcje to *liniowa*, *ReLU* (Rectified Linear Unit), *tanh* (hiperboliczny tangens) i *sigmoid*.

**Reglarization** pomaga zapobiegać przeuczeniu przez dodanie kary do funkcji strat za zbyt duże wagi. W TensorFlow Playground dostępne są dwa typy regularizacji: L1 i L2. L1 kara za duże wagi prowadzi do modeli bardziej rzadkich (z większą liczbą wag równych zero), podczas gdy L2 kara za duże wagi prowadzi do bardziej "gładkich" modeli bez dużej liczby wag bliskich zeru.

**Problem type** możesz wybrać między problemem klasyfikacji a regresji. Klasyfikacja jest używana, gdy przewidujemy etykietę z określonego zestawu klas, a regresja, gdy przewidujemy ciągłą wartość.



##Zestawy danych

Po lewej stronie interfejsu znajduje się panel pozwalający na wybieranie różnych zestawów danych oraz modyfikowanie ich charakterystyk.
Każdy zestaw danych reprezentuje unikalny wzór lub rozkład danych, który sieć neuronowa ma nauczyć się rozpoznawać i klasyfikować. Dostępne wzory to:

- **Gaussian**: Dane są rozmieszczone wokół jednego lub kilku centralnych punktów, tworząc wzory przypominające rozkład normalny.
- **Spiral**: Dane są rozmieszczone w kształcie spirali, co stanowi wyzwanie dla sieci neuronowej ze względu na ich skomplikowany, zakręcony kształt.
- **Circle**: Dane są rozmieszczone w formie koła lub kilku koncentrycznych kół.
- **Exclusive OR**: Dane układają się w wzór, który symuluje operację XOR, gdzie klasy są mieszane na przeciwległych narożnikach przestrzeni.

Po wyborze zestawu danych można dostosować charakterystyki za pomocą suwaków znajdujących się pod ikonami wzorów danych. Dostępne opcje modyfikacji to:

- **Noise**: Pozwala na dodanie hałasu do danych, co czyni problem klasyfikacji trudniejszym poprzez wprowadzenie losowości do rozkładu punktów. Wyższy poziom hałasu sprawia, że granice między klasami stają się bardziej niejednoznaczne.
- **Ratio of training to test data**: Umożliwia zmianę proporcji danych używanych do treningu sieci w stosunku do danych używanych do jej walidacji. Zmiana tego stosunku może wpłynąć na zdolność modelu do generalizacji na nowych danych.

##Funkcje wejściowe

Tuż obok panelu z zestawami danych znajdują się funkcje wejściowe, które będą używane przez sieć neuronową do nauki. Cechy te są podstawą dla modelu do rozpoznawania wzorców w danych i dokonywania predykcji. Do wyboru mamy:

- **X1, X2**: Podstawowe cechy reprezentujące poziome i pionowe położenie punktów danych. Są to najprostsze cechy, które można użyć w modelu.
- **X1^2, X2^2**: Kwadraty podstawowych cech. Ich użycie może pomóc w modelowaniu nieliniowości w danych.
- **X1*X2**: Iloczyn dwóch cech, co może być przydatne w rozpoznawaniu wzorców, gdzie interakcja między cechami jest istotna.
- **Sin(X1), Sin(X2)**: Sinusy cech, które mogą pomóc modelowi w radzeniu sobie z okresowymi wzorcami w danych.


##Warstwy ukryte





Na środku strony widać schemat sieci neuronowej, który pozwala na manipulowanie jej architekturą poprzez dodawanie lub usuwanie warstw ukrytych oraz regulowanie liczby neuronów w każdej z tych warstw. Dzięki temu mamy wgląd w to, jak głębokość sieci i liczba neuronów wpływają na zdolność modelu do uczenia się i generalizacji.

**Dodawanie i Usuwanie Warstw Ukrytych**
- **Dodawanie warstwy**: Możesz dodać warstwę ukrytą, klikając na przycisk z symbolem plusa (+). Każda dodatkowa warstwa ukryta zwiększa głębokość modelu, co potencjalnie pozwala na uczenie się bardziej skomplikowanych wzorców w danych. Jest to jednak równoznaczne z większą złożonością obliczeniową i ryzykiem przeuczenia, zwłaszcza przy małej ilości danych.
- **Usuwanie warstwy**: Możesz usunąć warstwę ukrytą, klikając na przycisk z symbolem minusa (-), który pojawia się obok każdej warstwy ukrytej. Usunięcie warstw zmniejsza złożoność modelu, co może być przydatne w walce z przeuczeniem.

**Regulowanie Liczby Neuronów**
- **Zwiększanie liczby neuronów**: W każdej warstwie ukrytej możesz zwiększyć liczbę neuronów, klikając na przycisk z symbolem plusa (+) znajdujący się obok reprezentacji warstwy. Dodanie neuronów do warstwy zwiększa zdolność modelu do rozpoznawania i modelowania bardziej złożonych relacji w danych.
- **Zmniejszanie liczby neuronów**: Możesz zmniejszyć liczbę neuronów, klikając na przycisk z symbolem minusa (-). Zmniejszenie liczby neuronów może pomóc w zmniejszeniu ryzyka przeuczenia przez uproszczenie modelu.

**Wpływ na Uczenie**
- **Złożoność modelu**: Dodawanie większej liczby warstw i neuronów zwiększa złożoność modelu, co może poprawić jego zdolność do nauki złożonych wzorców. Jednakże, zwiększa to także ryzyko przeuczenia, zwłaszcza gdy dostępne są ograniczone dane treningowe.
- **Ryzyko przeuczenia**: Zbyt wiele warstw lub neuronów w stosunku do złożoności problemu i rozmiaru dostępnego zestawu danych może prowadzić do przeuczenia. Oznacza to, że model może doskonale nauczyć się danych treningowych, ale będzie miał problemy z generalizacją na nowych, niewidzianych danych.
- **Czas i zasoby obliczeniowe**: Większa głębokość sieci i liczba neuronów wymagają większych zasobów obliczeniowych i czasu na trening.

**Wizualizacja Wyników**

Panel "Output" zawiera graficzną reprezentację przestrzeni cech, gdzie dane wejściowe są wizualizowane na dwuwymiarowej płaszczyźnie. Kolory w tej przestrzeni reprezentują różne klasy (w przypadku klasyfikacji) lub gradient przewidywanych wartości (w teoretycznym przypadku regresji).

**Interpretacja**

- **Granice Decyzyjne**: W przypadku klasyfikacji, granice decyzyjne pokazują, gdzie model "decyduje" o zmianie klasyfikacji z jednej klasy na inną. Są one bezpośrednim wynikiem funkcji aktywacji i struktury (głębokości i szerokości) sieci.
- **Regiony Klas**: Obszary o jednolitym kolorze reprezentują regiony przestrzeni cech, które model przewiduje jako należące do tej samej klasy. W idealnym przypadku, dane należące do tej samej klasy powinny być zawarte w jednym regionie, bez przeplatających się obszarów z innymi klasami.

**Metryki Training Loss i Test Loss**

- **Training Loss (Strata Treningowa)**
Miara błędu modelu na danych, na których był trenowany.
Niska strata wskazuje na dobre dopasowanie do danych treningowych. Bardzo niska wartość może jednak sygnalizować przeuczenie, tj. gdy model jest zbyt ściśle dopasowany do treningu i traci zdolność do generalizacji.

- **Test Loss (Strata Testowa)**
Miara błędu modelu na danych testowych, które nie były używane podczas treningu.
Niska strata testowa oznacza, że model dobrze generalizuje, czyli dokładnie przewiduje wyniki na nowych danych. Wysoka różnica między stratą treningową a testową może wskazywać na przeuczenie.
