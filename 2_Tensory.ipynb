**UWAGA**

Wektor 5-wymiarowy NIE jest tensorem 5-wymiarowym.
- 5-wymiarowy wektor ma 1 oś i 5 wymiarów (elementów) ułożonych wzdłuż osi
- 5- wymiarowy tensor ma 5 osi i dowolnie dużą liczbę wymiarów (elementów) wzdłuż każdej z nich

Załóżmy, że mamy zbiór danych składający się z 2500 książek. Każda książka jest opisana za pomocą wektora cech, który może zawierać 10 informacji o każdej z książek. Jaki będzie kształt tensora?


**dane szeregu czasowego i dane sekwencyjne:**
- kiedy dane tworzą uporządkowaną sekwencje, czyli jest w nich uwzględniony upływ czasu to najlepiej je zapisać jako tensory 3D z osią czasu


Przykład:

Zbiór danych - postów na portalu społecznościowym, każdy post jest sekwencją 280 znaków, każdy znak może być jednym ze znaków ze zbioru o 128 elementach. Każdy znak może być zakodowany w formie binarnego wektora o rozmiarze 128. Jaki kształt będzie miał tensor postu, a jaki zbiór miliona postów?

Przykład dotyczy techniki: "kodowania gorącojedynkowego" (ang. one-hot encoding).

Wektor binarny: Każdy unikalny znak z alfabetu jest reprezentowany jako 128-wymiarowy wektor binarny. W tym wektorze wszystkie pozycje są wypełnione zerami, z wyjątkiem jednej pozycji, która odpowiada danemu znakowi i jest oznaczona jako 1. Pozycja ta jest określona przez indeks znaku w alfabetcie.

Masz zbiór danych zawierający 5000 recenzji filmów. Każda recenzja jest reprezentowana jako sekwencja 200 słów, gdzie każde słowo jest zapisane jako wektor 50-wymiarowy. Przygotuj tensor, który będzie zawierał cały zbiór danych zgodnie z opisem. Określ kształt tensora i wytłumacz, co każdy z wymiarów reprezentuje.

**Dane w postaci obrazów:**

- obrazy najczęściej są trójwymiarowe - mają: wysokość, szerokość oraz głębię kolorów
- obrazy w skali szarości mają tylko jeden kanał koloru, dlatego można je zapisać jako tensory 2D, natomiast przyjmuje się, że obrazy są zawsze trójwymiarowe, a te monochromatyczne mają jednowymiarowy kanał koloru
- Wsad 128 obrazów w skali szarości o rozmiarze 256 x 256 mogłaby być więc przechowywana w tensorze o kształcie **(128, 256, 256, 1)**
- Wsad 128 kolorowych obrazów **(128, 256, 256, 3)**


Liczba **3** w przypadku kolorowych obrazów odnosi się do trzech kanałów kolorów w **modelu kolorów RGB**, który jest standardem w cyfrowym przetwarzaniu obrazów. Model RGB używa trzech kolorów światła: czerwonego (Red), zielonego (Green) i niebieskiego (Blue), do tworzenia pełnego spektrum kolorów widzianego przez ludzkie oko.

**Dane w postaci materiałów wideo:**
- wymagają zastosowania tensorów 5D
- Materiały wideo to tak naprawdę sekwencje klatek, każda klatka to po prostu kolorowy obraz


(wysokość, szerokość, głębokość koloru)  kolorowy obraz - tensor 3D

 (klatki, wysokość, szerokość, głębokość koloru) sekwencje klatet - tensor 4D

 (próbki, klatki, wysokość, szerokość, głębokość koloru) wsad różnych wideo - tensor 5D
